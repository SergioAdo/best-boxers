```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
box= pd.read_csv('box.csv')
box.head()
```

```python
box['w-l-d']= box['w-l-d'].apply(lambda x: x.replace('�', '-'))
box.head()
```

```python
box['win']= (box['w-l-d'].apply(lambda x : x.split('-')[0])).astype('int64')
box.head()
```

```python
box['losses']= (box['w-l-d'].apply(lambda x : x.split('-',2)[1])).astype('int64')
box.head()
```

```python
box['draws']= (box['w-l-d'].apply(lambda x : x.split('-',2)[-1])).astype('int64')
box.head()
```

```python
box['total']= box['win']+box['losses']+box['draws']
box.head()
```

```python
box['win_perc']= round((box['win']/box['total'])*100, 0)
box.head()
```

```python
box['nationality']= box['residence'].apply(lambda x: x.split(',')[-1])
box['nationality'].value_counts()
```

```python
box.head()
```

```python
box.columns
```

```python
box['career_start']= (box['career'].apply(lambda x : x.split('-')[0])).astype('int64')
box['career_end']= (box['career'].apply(lambda x : x.split('-')[-1])).astype('int64')
box['years_career']= box['career_end']-box['career_start']
box.head()
```

```python
df= box[['rank', 'name','division','years_career','stance', 'total','win', 'losses', 'draws','win_perc', 'nationality', 'residence']]
```

```python
df.to_csv('boxerstats.csv', index=False)
```
